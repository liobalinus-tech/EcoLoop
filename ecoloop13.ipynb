```python
# --- 1. INSTALL LIBRARIES ---
!pip install ultralytics roboflow

import os
import shutil
import json
import zipfile
from ultralytics import YOLO
from roboflow import Roboflow

# --- 2. YOUR SPECIFIC DATA SOURCE ---
print("📥 Fetching EcoLoop Master v2 data...")
rf = Roboflow(api_key="gC0a7I5Ubjo6OqyNENgv")
project = rf.workspace("thelioba").project("ecoloop_master_v2")
dataset = project.version(1).download("yolov8")

# --- 3. TRAINING (100 EPOCHS) ---
print("🏗️ Building the 100-epoch Production Model...")
model = YOLO('yolov8n.pt') 
model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=100,
    imgsz=640,
    patience=20, # Automatically stops if the model stops improving
    project='EcoLoop_Production',
    name='v2_100_epochs',
    save=True,
    device=0
)

# --- 4. CREATE CARBON MATH LOGIC ---
carbon_data = {
    "project": "EcoLoop",
    "version": "2.0",
    "factors": {
        "METAL": 2.5, "PLASTIC": 1.5, "WOOD": 2.0, 
        "CARDBOARD": 0.9, "GLASS": 0.5, "BIODEGRADABLE": 1.0
    }
}
with open('/kaggle/working/carbon_factors.json', 'w') as f:
    json.dump(carbon_data, f, indent=4)

# --- 5. BUNDLE FOR THE ENGINEERS ---
print("📦 Zipping results for handover...")
with zipfile.ZipFile('EcoLoop_Production_Package.zip', 'w') as zipf:
    # Add the Brain (Weights)
    best_weights = '/kaggle/working/EcoLoop_Production/v2_100_epochs/weights/best.pt'
    if os.path.exists(best_weights):
        zipf.write(best_weights, arcname='best.pt')
    
    # Add the Logic (Math)
    zipf.write('/kaggle/working/carbon_factors.json', arcname='carbon_factors.json')
    
    # Add the Success Proof (Chart)
    chart = '/kaggle/working/EcoLoop_Production/v2_100_epochs/results.png'
    if os.path.exists(chart):
        zipf.write(chart, arcname='accuracy_results.png')

print("✅ DONE! Everything is in 'EcoLoop_Production_Package.zip'")
```
